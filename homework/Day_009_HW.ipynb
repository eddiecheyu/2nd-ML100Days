- 模擬一組負相關的資料，並計算出相關係數以及畫出 scatter plot

```python
# 載入基礎套件
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#生成隨機的x和y（較不相關）
x = np.random.randint(0, 50, 500)
y = np.random.randint(0, 50, 500)
#印出相關係數矩陣
print(np.corrcoef(x, y))
#畫出散佈圖
plt.scatter(x, y)
```

```python
#生成x和y（相關係數較高的）
x = np.random.randint(0, 50, 500)
y = x + np.random.randint(0, 50, 500)
#印出相關係數矩陣
print(np.corrcoef(x, y))
#畫出散佈圖
plt.scatter(x, y)
```
