```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
plt.style.use('ggplot')

# 忽略警告訊息
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
#用 numpy 建立一個 10 x 10, 數值分布自 -1.0 ~ 1.0 的矩陣並繪製 Heatmap
"""
建立隨機0-1之間的小數矩陣為np.random.random((x,x))
因為這題要建立-1~1之間，要透過a + (b-a) * np.random.random((x, x))這個算式
a和b就是a-b之間
Your Code Here
"""
matrix = -1 + (1-(-1))*np.random.random((10,10))

plt.figure(figsize=(10,10))
"""
Your Code Here
用matplotlib做heatmap很麻煩，直接用seaborn的圖來做
"""
heatmap = sns.heatmap(matrix)

plt.show()
```

```python
#用 numpy 建立一個 1000 x 3, 數值分布為 -1.0 ~ 1.0 的矩陣，並繪製 PairPlot (上半部為 scatter, 對角線為 hist, 下半部為 density)
nrow = 1000
ncol = 3
"""
Your Code Here
"""
matrix = -1 + (1-(-1))*np.random.random((nrow, ncol))

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice)

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)
"""
Your Code Here
Please replace "..." to correct plot function
第一個是散佈圖，第二個是直方圖，第三個是密度圖（plt沒有，用seaborn的圖）
"""
grid.map_upper(plt.scatter , alpha = 0.2)
grid.map_diag(plt.hist )
grid.map_lower(sns.kdeplot , cmap = plt.cm.OrRd_r)

plt.show()
```

```python
#用 numpy 建立一個 1000 x 3, 數值分布為常態分配的矩陣，並繪製 PairPlot (上半部為 scatter, 對角線為 hist, 下半部為 density)
nrow = 1000
ncol = 3
"""
Your Code Here
"""
matrix = np.random.normal(size=(nrow, ncol))
                          
# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice)

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)
"""
Your Code Here
Please replace "..." to correct plot function
"""
grid.map_upper(plt.scatter , alpha = 0.2)
grid.map_diag(plt.hist )
grid.map_lower(sns.kdeplot , cmap = plt.cm.OrRd_r)

plt.show()
```
