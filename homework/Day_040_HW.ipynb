試著使用 sklearn datasets 的其他資料集 (boston, ...)，來訓練自己的線性迴歸模型，並加上適當的正則話來觀察訓練情形。

```python
#import package
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.linear_model import Ridge
```

```python
from sklearn import datasets
boston = datasets.load_boston()
df = pd.DataFrame(data=boston['data'], columns = boston['feature_names'])
print(df.shape)
df.head(10)
```

```python
Y= boston['target']
Y
```

```python
#先用一般線性回歸做做看
train_x, test_x, train_y, test_y = train_test_split(df, Y, test_size=0.3)
lr = LinearRegression()
lr.fit(train_x, train_y)
pred_y = lr.predict(test_x)
```

```python
#可以看回歸模型的參數值
print('Coefficients: ', lr.coef_)

#預測值與實際值的差距，使用 MSE
print("Mean squared error: %.2f"% mean_squared_error(test_y, pred_y))

#看測試和預測的y的r平方
print('R Squared score:', r2_score(test_y, pred_y))

#看測試和預測的y相關係數
print('Correlation of coefficient:', stats.pearsonr(test_y, pred_y)[0])
```

```python
#使用Ridge調整alpha
for a in [1e-3, 1e-2, 1e-1, 1, 1e1, 1e2, 1e3]:
    rid = Ridge(alpha=a)
    rid.fit(train_x, train_y)
    pred_y = rid.predict(test_x)
    print('Alpha:', a)
    print('Coefficients: ', rid.coef_)
    print("Mean squared error: %.2f"% mean_squared_error(test_y, pred_y))
    print('R Squared score:', r2_score(test_y, pred_y))
    print('Correlation of coefficient:', stats.pearsonr(test_y, pred_y)[0])
```

```python
#使用Lasso調整alpha
for a in [1e-3, 1e-2, 1e-1, 1, 1e1, 1e2, 1e3]:
    ls = Lasso(alpha=a)
    ls.fit(train_x, train_y)
    pred_y = ls.predict(test_x)
    print('Alpha:', a)
    print('Coefficients: ', ls.coef_)
    print("Mean squared error: %.2f"% mean_squared_error(test_y, pred_y))
    print('R Squared score:', r2_score(test_y, pred_y))
    print('Correlation of coefficient:', stats.pearsonr(test_y, pred_y)[0])
```

使用Lasso因為會將係數降至0，所以alpha值不能設定過大，會導致模型沒有特徵能夠擬合資料

使用Ridge會將係數隨著alpha值慢慢趨近於0

在這個資料中，是否有正規化對模型的擬合和預測沒有太大的影響
