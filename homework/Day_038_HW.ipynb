```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
```

# Linear Regression

```python
from sklearn import datasets
boston = datasets.load_boston()
boston
```

```python
boston_data = pd.DataFrame(data=boston['data'], columns = boston['feature_names'])
print(boston_data.shape)
boston_data.head(10)
```

```python
boston_y = boston['target']
boston_y
```

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(boston_data, boston_y, test_size=0.3)
```

```python
reg = LinearRegression().fit(train_x, train_y)
pred_y = reg.predict(test_x)
```

```python
#可以看回歸模型的參數值
print('Coefficients: ', reg.coef_)

#預測值與實際值的差距，使用 MSE
from sklearn.metrics import mean_squared_error
print("Mean squared error: %.2f"% mean_squared_error(test_y, pred_y))

#看測試和預測的y相關係數
import scipy.stats as stats
print('Correlation of coefficient:', stats.pearsonr(test_y, pred_y)[0])
```

# Logistic Regression

```python
wine = datasets.load_wine()
wine
```

```python
wine_df = pd.DataFrame(data=wine['data'], columns=wine['feature_names'])
target = wine['target']
print(wine_df.shape)
wine_df.head(10)
```

```python
train_x, test_x, train_y, test_y = train_test_split(wine_df, target, test_size=0.4)
```

```python
model = LogisticRegression().fit(train_x, train_y)
pred_y = model.predict(test_x)
pred_y
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(test_y, pred_y)
print("Accuracy: ", acc)
```
